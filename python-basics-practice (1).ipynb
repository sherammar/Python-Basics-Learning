```python
# Question: Write a function is_palindrome(word) that returns True if the word is the same forward and backward.

#ğŸ‘‰ A palindrome means the word reads the same forwards and backwards (like madam, level, racecar).

#                -----My personel approach-----
# def is_palindrome(word):
#     if word == word[::-1]:  # Reversing the string and check the equality by Boolean(True, False)
#         return True
#     else:
#         return False
 
# print("Given Word is palindrome? ",is_palindrome("level"))

# Alternative approach, much shorter

def is_palindrome(word):
    return word == word[::-1]

print("Given word is palindrome? ", is_palindrome("Ammar"))
print("Given word is palindrome? ", is_palindrome("racecar"))
```

```python
# Question: Write a function max_of_three(a, b, c) that returns the largest of three numbers.

# Using the max function

def max_of_three(a, b, c):
    largest = max(a, b, c)
    return largest

print(max_of_three(10,14,2))

# alternative code

def max_of_three_alt(a, b, c):
    return max(a, b, c)
print(max_of_three_alt(1,50,40))

# Alternative approach, using conditionals

def max_of_three_alt1(x, y, z):
    if x>y and x>z:
        print("Greatest number is: ", x)
    elif y>x and y>z:
        print("Greatest number is: ", y)
    else:
        print("Greatest number is: ", z)
        
max_of_three_alt1(200,50,300)
```

```python
# Question: Write a function calculate_discount(price, discount_percent) that returns the final price after discount.

def calculate_discount(price, discount_percent):

    # To calculate price after discount, first we divide the percentage/100,
    # then multiply original price by decimal and then subtract the value from original price
    percentage = discount_percent / 100
    total_discount = price * percentage
    final_price = price - total_discount
    return final_price
    
calculate_discount(200,15)
```

```python
# Question: Write a function is_even_length(text) that returns True if the string length is even.

def is_even_length(text):
    if len(text) % 2 == 0:   # finding modulus(remainder) of text and comparing with zero
        return True
    else:
        return False
is_even_length("ammar")
```
